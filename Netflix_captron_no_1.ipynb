# Visualization

```python
df.type.value_counts()  # value_counts method show the count of different categories
```

```python
ax = sns.countplot(x = "type", data = df)  # count plot help us plot counts of each categories

for bars in ax.containers:
  ax.bar_label(bars)
```

# country analysis

```python
df['country'].value_counts().head(10)  #total movies and Tv showes
```

```python
plt.figure(figsize = (15,6))
ax = sns.countplot(y = 'country', order = df['country'].value_counts().index[0:10], data = df)
plt.title(' Top 10 countries_Movies')

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
movie_country = df[df['type'] == 'Movie']
tv_show_country = df[df['type'] == 'TV Show']
```

```python
plt.figure(figsize = (15,6))
ax = sns.countplot(y = 'country', order = df['country'].value_counts().index[0:10], data = movie_country)
plt.title(' Top 10 countries_Movies')

for bars in ax.containers:
  ax.bar_label(bars)



plt.figure(figsize = (15,6))
ax = sns.countplot(y = 'country', order = df['country'].value_counts().index[0:10], data = tv_show_country)
plt.title('Top 10 countries_tv_show')


for bars in ax.containers:
  ax.bar_label(bars)
```

```python
df.rating.value_counts()
```
